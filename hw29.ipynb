```python
import pandas as pd
from IPython.display import display
sports = pd.read_csv(r'sport_athletes.csv')
#1

display(sports)
display(sports.groupby("Sport Type")[['Weight(kg)', 'Height(cm)', 'Wins']].max().rename(columns={'Weight(kg)':"max_weight", 'Height(cm)':"max_height", 'Wins':"max_wins"}))
```

```python
#2
display(sports.groupby("Sport Type")[['Weight(kg)', 'Height(cm)', 'Wins']].min().sort_values(by='Weight(kg)', ascending = True))
```

```python
#3
display(sports[["Sport Type", "Athlete Name", 'Height(cm)']].sort_values(by="Height(cm)", ascending = False).groupby("Sport Type").head(2))
```

```python
#4
#a
display(sports.groupby(["Sport Type", "Country"]).max())

#b
display(sports.groupby(["Sport Type", "Country"]).max().head(1))

#c
display(sports.groupby(["Sport Type", "Country"]).max().xs('Soccer'))

#d
display(sports.groupby(["Sport Type", "Country"]).max().loc[['Soccer', 'Tennis']])

#e
display(sports.groupby(["Sport Type", "Country"]).max().swaplevel())

#f
display(sports.groupby(["Sport Type", "Country"]).agg({"Wins":['max', 'mean', 'min', 'std'], "Height(cm)":['max', 'mean', 'min', 'std']}))
```

```python
#5
new_rows = pd.DataFrame({ 'Country': ['Canada', 'USA'], 'Sport Type': ['Hockey', 'Skiing'],
'Athlete Name': ['Wayne Gretzky', 'Lindsey Vonn'], 'Height(cm)': [183, 172],
'Weight(kg)': [84, 68], 'Wins': [3, 4], 'Olympic Date': ['01/02/2002', '12/02/2010'] })
new_sports = pd.concat([sports, new_rows], axis = 0).reset_index()

```
